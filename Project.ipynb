## 1. Determine the distribution of employees across each team

```python
total_team=df.groupby('Team')['Name'].count()
total_employees = df['Name'].count()
total_team
sorted_df = total_team.sort_values(ascending=False)
sorted_df
```

# Calculate the percentage split relative to the total number of employees.

```python

Percentage= (total_team/ total_employees) * 100
Percentage
```

# Graphical Representation:Percentage of Distribution of Employees Across each Team by using lineplot

```python
plt.figure(figsize=(12,4))
plt.plot(Percentage.index, Percentage.values)
plt.xlabel('Team')
plt.ylabel('Percentage')
plt.title('Percentage of Distribution of Employees Across each Team')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

## 2. Segregate employees based on their positions

```python

position = df.groupby('Position')['Name'].count()
position
```

# Graphical Representation:Distribution of Employees by Position using pie plot

```python

plt.pie(position,labels=position.index,autopct='%0.1f%%')

plt.title('Distribution of Employees by Position')
plt.show()
```

# 3. Identify the predominant age group among employees

```python

age_distribution = df['Age'].value_counts()
age_distribution
```

# Graphical Representation:Distribution of Employees by Age using barplot

```python

plt.bar(age_distribution.index,age_distribution.values)
plt.xticks(range(19,41))
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Employees by Age')
plt.show()
```

## 4. Discover which team and position have the highest salary expenditure

```python
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
salary_expenditure 
```

```python
max_salary_expenditure = salary_expenditure.loc[salary_expenditure['Salary'].idxmax()]
max_salary_expenditure
```

# Graphical Representation:Salary Expenditure by Team and Position

```python
plt.figure(figsize=(10, 4))
plt.bar(salary_expenditure['Team'], salary_expenditure['Salary'])
plt.bar(salary_expenditure['Position'], salary_expenditure['Salary'])
plt.xticks(rotation=90)
plt.xlabel('Team - Position')
plt.ylabel('Salary Expenditure')
plt.title('Salary Expenditure by Team and Position')
plt.grid()
plt.show()
```

## 5. Investigate if there's any correlation between age and salary

```python
cor = df['Age'].corr(df['Salary'])
cor  
```

```python
plt.bar(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation between Age and Salary')
plt.xticks(range(18,41))
plt.show()  
```
